```python
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.qualite_eau")
print("✅ Schema workspace.qualite_eau créé")
```

```python
volumes = ["raw", "bronze", "silver", "gold"]
for v in volumes:
    spark.sql(f"CREATE VOLUME IF NOT EXISTS workspace.qualite_eau.{v}")
    print(f"✅ Volume '{v}' créé")
```

```python
display(spark.sql("SHOW VOLUMES IN workspace.qualite_eau"))
```

```python
CATALOG = "workspace"
DATABASE = "qualite_eau"

RAW_PATH    = f"/Volumes/{CATALOG}/{DATABASE}/raw"
BRONZE_PATH = f"/Volumes/{CATALOG}/{DATABASE}/bronze"
SILVER_PATH = f"/Volumes/{CATALOG}/{DATABASE}/silver"
GOLD_PATH   = f"/Volumes/{CATALOG}/{DATABASE}/gold"

print("Chemins configurés :")
print(f"  RAW    : {RAW_PATH}")
print(f"  BRONZE : {BRONZE_PATH}")
print(f"  SILVER : {SILVER_PATH}")
print(f"  GOLD   : {GOLD_PATH}")
```

```python
import os

RAW_PATH = "/Volumes/workspace/qualite_eau/raw"

for dept in sorted(os.listdir(RAW_PATH)):
    dept_path = f"{RAW_PATH}/{dept}"
    if os.path.isdir(dept_path):
        fichiers = os.listdir(dept_path)
        print(f"📁 {dept} : {len(fichiers)} fichier(s)")
        for f in fichiers:
            taille = os.path.getsize(f"{dept_path}/{f}")
            print(f"   - {f} ({taille/1024/1024:.1f} MB)")
```
